```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import random
import time

base_urls = [
    'https://www.zillow.com/professionals/real-estate-agent-reviews/punta-gorda-fl/',
    'https://www.zillow.com/professionals/real-estate-agent-reviews/oshkosh-wi/',
    'https://www.zillow.com/professionals/real-estate-agent-reviews/philadelphia-pa/',
    'https://www.zillow.com/professionals/real-estate-agent-reviews/nashville-tn/',
    'https://www.zillow.com/professionals/real-estate-agent-reviews/denver-co/',
    'https://www.zillow.com/professionals/real-estate-agent-reviews/boston-ma/',
    'https://www.zillow.com/professionals/real-estate-agent-reviews/cleveland-oh/',
    'https://www.zillow.com/professionals/real-estate-agent-reviews/houston-tx/',
    'https://www.zillow.com/professionals/real-estate-agent-reviews/phoenix-az/',
    'https://www.zillow.com/professionals/real-estate-agent-reviews/san-antonio-tx/',
    'https://www.zillow.com/professionals/real-estate-agent-reviews/hollywood-fl/'
]

headers = {
    'User-Agent': '',  # dynamic user agents
    'Accept-Language': 'en-US,en;q=0.9,ar;q=0.8,es;q=0.7,tr;q=0.6',
    'Cache-Control': 'max-age=0',
    'DNT': '1',
    'Sec-Fetch-Dest': 'document',
    'Sec-Fetch-Mode': 'navigate',
    'Sec-Fetch-Site': 'same-origin',
    'Sec-Fetch-User': '?1',
    'Upgrade-Insecure-Requests': '1',
}

agent_data = []
user_agents = [
    'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/92.0.4515.107 Safari/537.36',
    'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36',
]

session = requests.Session()

for base_url in base_urls:
    state = base_url.split('/')[-2].replace('-', ' ').upper()

    for page in range(1, 26):
        url = base_url if page == 1 else f'{base_url}?page={page}'
        headers['User-Agent'] = random.choice(user_agents)
        response = session.get(url, headers=headers)
        print(f"Response Code for {url}: {response.status_code}")

        if response.status_code == 200:
            html = response.text
            soup = BeautifulSoup(html, 'html.parser')
            table = soup.find('table', attrs={'aria-label': "Agent's table"})
            agent_body = table.find('tbody')
            agent_rows = agent_body.find_all('tr')

            for row in agent_rows:
                agent_info = row.find_all('td')[0]
                agent_name = agent_info.find('a')
                agent_name_text = agent_info.find('a').text
                agent_pi = agent_name.find_parent('div')
                agent_number_divs = agent_pi.find_all('div')

                if len(agent_number_divs) > 0:
                    agent_number = agent_number_divs[0].text.replace('phone number', '').strip()
                else:
                    agent_number = 'N/A'  # Placeholder value for agents without a phone number

                agent_data.append({'Agent Name': agent_name_text, 'Agent Number': agent_number, 'State': state})

        # Introduce a delay between requests
        delay = random.uniform(2, 5)
        time.sleep(delay)
```

```python
df = pd.DataFrame(agent_data)
df
```

```python
df.to_csv('agents.csv', index=False)
```
